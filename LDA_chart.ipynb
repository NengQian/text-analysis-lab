```python
import json
from pprint import pprint

import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
```

```python
#FILE = 'bmw2013_money.json'
FILE = 'bmw2013_money.json'
```

```python
#input file
with open(FILE) as f:
    data = json.load(f)

para_num = []
related = []
    
for item in data:
    para_num.append(item['paragraph'])
    related.append(float(item['value']))
```

```python
para_num.index(714)
```

## pie chart

```python
#choose range and do the normalization
para = para_num[0:10]
rela = related[0:10]

d = sum(rela)
rela = [x/d for x in rela]
```

```python
#draw the pie chart
labels = para_num
values = rela

trace = go.Pie(labels=labels, values=values)

#py.iplot([trace], filename='basic_pie_chart')
```

```python
py.offline.iplot([trace], filename='basic_pie_chart')
```

## bar chart

```python
#build a new sorted list
bar_para=list(range(0,1100))
bar_rela=list(range(0,1100))
```

```python
#print(bar_para)
```

```python
for item in bar_para:
    if item in para_num:
        #print(item)
        #print(para_num.index(item))
        bar_rela[item] = related[para_num.index(item)]
        #print(bar_para[item])
        #print(item)
        #bar_para[item]
    else:
        bar_rela[item] = 0
```

```python
#bar_rela
```

```python
#adjust the list
bar_para = bar_para[1:]
bar_rela = bar_rela[1:]
```

```python
#draw the bar chart
trace = go.Bar(x=bar_para, y=bar_rela)
```

```python
py.offline.iplot([trace], filename='money topic in BMW2013')
```
